```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Generate synthetic data
np.random.seed(0)
num_samples = 1000

age = np.random.randint(18, 80, num_samples)
income = np.random.normal(50000, 15000, num_samples)
expenditure = np.random.normal(30000, 10000, num_samples)
```

```python
# Create a DataFrame
df = pd.DataFrame({
    'Age': age,
    'Income': income,
    'Expenditure': expenditure
})
```

```python
# Histogram
plt.figure(figsize=(10, 6))
plt.hist(df['Age'], bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
# Bar graph
plt.figure(figsize=(10, 6))
df_income_bins = pd.cut(df['Income'], bins=5)
income_groups = df.groupby(df_income_bins)['Expenditure'].mean()
income_groups.plot(kind='bar', color='orange', alpha=0.7)
plt.title('Mean Expenditure by Income Groups')
plt.xlabel('Income Groups')
plt.ylabel('Mean Expenditure')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Income', y='Expenditure', data=df, alpha=0.5)
plt.title('Scatter Plot of Income vs Expenditure')
plt.xlabel('Income')
plt.ylabel('Expenditure')
plt.grid(True)
plt.show()
```

```python
# Heatmap
plt.figure(figsize=(10, 6))
heatmap_data = df.pivot_table(index='Age', columns='Income', values='Expenditure', aggfunc=np.mean)
sns.heatmap(heatmap_data, cmap='YlGnBu')
plt.title('Heatmap of Age vs Income vs Expenditure')
plt.xlabel('Income')
plt.ylabel('Age')
plt.show()
```
